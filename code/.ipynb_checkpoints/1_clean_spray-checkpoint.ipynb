### Clean Data

```python
#spray['latitude'] = spray.Latitude.apply(int)
#spray['longitude'] = spray.Longitude.apply(int)

def create_yr(x): 
    return x.split('-')[0] 

def create_mth(x): 
    return x.split('-')[1] 

def create_day(x): 
    return x.split('-')[2] 

def rename_columns (columns):
    return [column.lower() for column in columns]

def clean_data(df): 
    df['year_spray'] = df.Date.apply(create_yr)
    df['month_spray'] = df.Date.apply(create_mth)
    df['day_spray'] = df.Date.apply(create_day)
    #df['latitude'] = df.Latitude.apply(int)
    #df['longitude'] = df.Longitude.apply(int)
    df.drop(['Date', 'Time'], axis = 1, inplace = True)
    df.columns = rename_columns(df.columns)
    return df

spray = clean_data(spray)
```
